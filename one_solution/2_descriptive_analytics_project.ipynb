## Quick Recap:
- There are 4 types of Aggregations: 

    1. **Data Summarization**: sum, count, min, max

    2. **Grouping Data**: categorizing on characteristics 

    3. **Time-based Aggregation**: consolidating data over various time intervals

    4. **Hierarchical Aggregation**: aggregating data at different levels of a hierrachical structure

        Remember that the questions or ideas that you want to explore may involve a combination of these types of aggregations.

- The Measures of Central Tendency: mean, median, mode

- The Measures of Dispersion: range, IQR, variance, Standard Deviation

- Measure of shape: skewness

#### Displaying the percentage values of total sales per region

````python
# sorting the values
sales_region_sorted = sales_region.sort_values(by="sales",ascending=False)
````

````python
ax = sns.barplot(x="region", y="sales", data=sales_region_sorted)

# Calculate the percentage of sales for each region
total_sales = sales_region_sorted["sales"].sum()

#ax.patches represents a collection of patches (bars) in the barplot.
for p in ax.patches:
    #For each patch, we retrieve the height of the bar using p.get_height()
    height = p.get_height()
    
    #p.get_x() + p.get_width() / 2 calculates the x-coordinate position for the text label to be placed at the center of the bar.
    ax.text(p.get_x() + p.get_width() / 2, height + 5, f"{(height/total_sales)*100:.2f}%", ha="center")

# Customize the plot
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales Percentage")

# Display the plot
plt.show()
````
